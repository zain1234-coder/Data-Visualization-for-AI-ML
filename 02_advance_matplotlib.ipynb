# 📊 Matplotlib 3D Plots

## Introduction

Most visualizations are two-dimensional (2D), where data is represented using only the **X-axis** and **Y-axis**.

However, many real-world datasets contain three variables. In such cases, **3D plots** allow us to visualize relationships using:

- X-axis
- Y-axis
- Z-axis

Matplotlib provides built-in support for creating 3D visualizations through the **mpl_toolkits.mplot3d** module.

---

# Why Use 3D Plots?

3D plots are useful when:

- Visualizing three variables simultaneously.
- Understanding complex mathematical functions.
- Exploring scientific and engineering data.
- Visualizing machine learning surfaces.
- Displaying terrain and geographical data.

---

# Import Required Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

---

# Creating a 3D Figure

Every 3D plot starts by creating a figure and adding a 3D axis.

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(111, projection='3d')
```

### Explanation

- `figure()` creates a new figure.
- `add_subplot()` adds an axis.
- `projection='3d'` converts it into a 3D axis.

---

# 1. 3D Line Plot

A 3D line plot displays a continuous line in three-dimensional space.

## Example

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

x = np.arange(0,10,0.2)
y = np.sin(x)
z = np.cos(x)

ax.plot(x,y,z,color='blue')

ax.set_title("3D Line Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

### Output

A spiral-like line showing how X, Y, and Z change together.

---

# 3. 3D Surface Plot

Surface plots visualize mathematical surfaces.

## Example

```python
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)

X,Y = np.meshgrid(x,y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')

ax.plot_surface(
    X,
    Y,
    Z,
    cmap='viridis'
)

plt.show()
```

---

# What is meshgrid()?

`meshgrid()` converts two 1D arrays into coordinate matrices.

Without it, Matplotlib cannot create surface plots.

Example

```python
x = np.arange(3)
y = np.arange(3)

X,Y = np.meshgrid(x,y)
```

Result

```
X

[[0 1 2]
 [0 1 2]
 [0 1 2]]

Y

[[0 0 0]
 [1 1 1]
 [2 2 2]]
```

# 6. 3D Bar Plot

A 3D version of a bar chart.

```python
x = np.arange(5)
y = np.arange(5)

z = np.zeros(5)

dx = np.ones(5)
dy = np.ones(5)

dz = [3,5,2,6,4]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')

ax.bar3d(
    x,
    y,
    z,
    dx,
    dy,
    dz,
    color='skyblue'
)

plt.show()
```

Applications

- Sales comparison
- Inventory analysis
- Business dashboards

# 🎨 Matplotlib Color Encoded Graphs

## Introduction

In many datasets, displaying only the X and Y coordinates is not enough to understand the complete information.

A **Color Encoded Graph** uses **colors** to represent an additional variable. This means a single visualization can display **three or even four dimensions of data**:

- X-axis
- Y-axis
- Color
- (Optional) Size

Color encoding helps identify patterns, trends, clusters, and outliers more effectively than a simple graph.

---

# Why Use Color Encoding?

Color encoding is useful when you want to visualize an extra numerical variable without creating another graph.

It helps in:

- Showing intensity
- Representing density
- Comparing values
- Detecting clusters
- Highlighting patterns
- Visualizing machine learning features

---

# Required Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

---

# Basic Color Encoded Scatter Plot

The most common color encoded graph is the Scatter Plot.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = np.random.rand(100)

colors = np.random.rand(100)

plt.figure(figsize=(8,6))

plt.scatter(
    x,
    y,
    c=colors,
    cmap='viridis'
)

plt.colorbar()

plt.title("Color Encoded Scatter Plot")

plt.show()
```

---

## Explanation

Here,

```
c=colors
```

tells Matplotlib to color every point according to the values stored in `colors`.

```
cmap='viridis'
```

selects the color palette.

```
plt.colorbar()
```

adds a color scale to explain what each color represents.

---

# What is a Colormap?

A **Colormap** is a collection of colors used to represent numerical values.

# 3d Surface plot

````python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
````

````python
xx,yy = np.meshgrid(x,y) #meshgrid form grid
````

````python
z = xx**2 + yy**2 # calculating 
````

````python
fig  = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap='winter')
ax.set_title('z = x^2 + y^2')
fig.colorbar(p)
fig.savefig('Surface.png')
````

````python
fig  = plt.figure(figsize=(12,8))
z = np.sin(xx) + yy + np.cos(yy)

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap='rainbow')
ax.set_title('z = np.sin(xx) + yy + np.cos(yy)')
fig.colorbar(p)
````

# contour
