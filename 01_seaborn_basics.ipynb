### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

`````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
`````

`````python
tips = sns.load_dataset('tips')
tips.head(2)
`````

`````python
#sctterplot -> Axes level
sns.scatterplot(data=tips,x = 'total_bill',y = 'tip',hue='sex',style='time',size='size') #tell which represent  male or female
`````

`````python
# replot scatterplot -> Figure level  -> square shape
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size') # size size increase the sie of marker according to there quantity
`````

`````python
# if continuty needed then use line plot

df = px.data.gapminder()
df
`````

`````python
temp_df = df[df['country'] == 'Pakistan'] 
temp_df
`````

`````python
# line plot Axis level

sns.lineplot(data=temp_df , x='year',y='lifeExp')
`````

`````python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
`````

`````python
# hue -> style
temp = df[df['country'].isin(['Brazil','Pakistan','Germany'])]
temp.head(1)
`````

`````python
sns.relplot(data=temp,x='year',y='lifeExp',hue='country',kind='line',style='continent')# style tell which contienet they belong
`````

`````python
# facet plot 
 # facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex', row='day')
sns.relplot(data=tips , x='total_bill' , y='tip' , col='sex', kind='scatter') # col = make plots side by ide or row= form plot uper down 
`````

`````python
# col wrap

sns.relplot(data=df,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3) #for controlling amount of columns in one row
`````

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

`````python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
`````

`````python
# plotting univariate histogram
sns.histplot(data=tips, x='total_bill')
`````

`````python
# Using displot

sns.displot(data=tips, x='total_bill',kind='hist') 
`````

`````python
# bins
sns.displot(data=tips, x='total_bill',kind='hist',bins=4) # bins control bins 
`````

`````python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# countplot  # histogram acting like countplot
sns.displot(data=tips, x='day', kind='hist' , ) 
`````

`````python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist' , hue='sex',element='step') 
`````

`````python
titanic = sns.load_dataset('titanic')
titanic
`````

`````python
sns.displot(data=titanic,x='age',kind='hist',hue='sex')
`````

`````python
# facet cols and rows
sns.displot(data=titanic,x='age',kind='hist',col='sex')
`````

`````python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=titanic,x='survived',hue='sex')
`````

`````python
# hue = Fill
sns.kdeplot(data=titanic,x='survived',hue='sex',fill=True) # fill filed it 
`````

`````python
# figure plot 

sns.displot(data=titanic,x='survived',hue='sex',fill=True,kind='kde')
`````

## Rugplot

### Plot marginal distributions by drawing ticks along the x and y axes.

#### This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

`````python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill') # tells density
`````

## Bivariate histogram
### A bivariate histogram bins the data within rectangles that tile the plot 
#### and then shows the count of observations within each rectangle with the fill color
````
sns.histplot(data=tips, x='total_bill', y='tip')

`````python
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
`````

# Bivariate Kdeplot
## a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

`````python
sns.displot(data=titanic,x='survived',y='pclass',kind='kde')
`````

### 2. Matrix Plot

- Heatmap
- Clustermap
  
## Heatmap
## Plot rectangular data as a color-encoded matrix

`````python
temp__df = df.pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(10,5))
sns.heatmap(temp__df) # don't have any  figure level function it can be done only by using axis level func
`````

`````python
# Same Graph for euorpio
plt.figure(figsize=(10,9))
plt.title('Life_Expectancies of European Countries')
sns.heatmap(df[df['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp'),annot=True,linewidths=0.5,cmap='rainbow')#linewidths for gaps
`````

# Clustermap

### Plot a matrix dataset as a hierarchically-clustered heatmap.

#### This function requires scipy to be available.

`````python
iris = px.data.iris()
iris
`````

`````python
sns.clustermap(iris.iloc[:,0:4])
`````
