## What is Matplotlib?

**Matplotlib** is one of the most popular Python libraries for data visualization. It allows you to create high-quality graphs, charts, and plots to analyze and present data.

It is widely used in:

# Basic Plot Structure

Every Matplotlib graph follows these steps:

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.title("Sample Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

# 1. Line Graph

A line graph is used to show trends over time.

Example

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]
sales = [15,18,25,22,30]

plt.figure(figsize=(8,5))

plt.plot(
    months,
    sales,
    color="blue",
    marker="o",
    linewidth=2,
    label="Sales"
)

plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.grid(True)
plt.legend()

plt.show()
```

---

# 2. Bar Graph

A bar graph compares different categories.

Example

```python
import matplotlib.pyplot as plt

students = ["Ali","Ahmed","Sara","John"]

marks = [85,92,78,95]

plt.figure(figsize=(8,5))

plt.bar(
    students,
    marks,
    color="skyblue",
    edgecolor="black"
)

plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

---

# 3. Horizontal Bar Graph

Horizontal bars are useful when category names are long.

Example

```python

plt.figure(figsize=(8,5))

plt.barh(

    edgecolor="black"
)

plt.title("Student Marks")

plt.xlabel("Marks")
plt.ylabel("Students")

```

---

# 4. Scatter Plot

Scatter plots show the relationship between two numerical variables.

Example

```python
import matplotlib.pyplot as plt

height = [150,160,165,170,180]

weight = [45,55,60,65,80]

plt.figure(figsize=(8,5))

plt.scatter(
    height,
    weight,
    color="red",
    s=100
)

plt.title("Height vs Weight")

plt.xlabel("Height")
plt.ylabel("Weight")

plt.grid(True)

plt.show()
```

---

# 5. Pie Chart

Pie charts display proportions of a whole.

Example

```python
import matplotlib.pyplot as plt

languages = ["Python","Java","C++","JavaScript"]

students = [45,20,15,20]

plt.figure(figsize=(8,8))

plt.pie(
    students,
    labels=languages,
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    explode=[0.05,0,0,0],
    wedgeprops={
        "edgecolor":"white",
        "linewidth":2
    }
)

plt.title("Programming Language Popularity")

plt.show()
```

# Best Practices

- Always add a title.
- Label both axes.
- Use meaningful colors.
- Keep charts simple.
- Add legends when plotting multiple datasets.
- Use grid lines when they improve readability.
- Choose the right chart for your data.
- Use `plt.tight_layout()` to avoid overlapping elements.

# Summary

Matplotlib provides several basic chart types:

- ✅ Line Graph → Trends over time
- ✅ Bar Graph → Category comparison
- ✅ Horizontal Bar Graph → Compare long category names
- ✅ Scatter Plot → Relationship between variables
- ✅ Pie Chart → Show percentages and proportions

These five chart types form the foundation of data visualization in Python and are essential for data analysis, machine learning, and business reporting.

### Bar chart

image.png

- Bivariate Analysis

````python
# simple bar 

x = [10,20,40,10,30]
y = ['red','blue','green','yellow','pink']

plt.bar(y,x , color=['red','blue','green','yellow','pink'])
````

````python
# multiple bar chat
batsman_rec = pd.read_csv('batsman_season_record.csv')
plt.bar(batsman_rec['batsman'],batsman_rec['2015'],width=0.5)
````

````python
plt.bar(np.arange(batsman_rec.shape[0]) - 0.2,batsman_rec['2015'],width=0.2,color='red')
plt.bar(np.arange(batsman_rec.shape[0]),batsman_rec['2016'],width=0.2,color='blue')
plt.bar(np.arange(batsman_rec.shape[0]) + 0.2,batsman_rec['2017'],width=0.2,color='green')
plt.xticks(np.arange(batsman_rec.shape[0]),batsman_rec['batsman'])
plt.show()
````

````python
# a problem if  nme is larger

x = [10,20,40,10,30]
y = ['red red red red red red red red',' blue blue blue blue blue blue','green green  green green green green','yellow','pink']

plt.bar(y,x , color=['red','blue','green','yellow','pink'])
plt.show()
````

````python
# solution
x = [10,20,40,10,30]
y = ['red red red red red red red red',' blue blue blue blue blue blue','green green  green green green green','yellow','pink']

plt.bar(y,x , color=['red','blue','green','yellow','pink'])
plt.xticks(rotation='vertical')
plt.show()
````

````python
# stack bar chat 
plt.bar(batsman_rec['batsman'],batsman_rec['2017'],label='2017')
plt.bar(batsman_rec['batsman'],batsman_rec['2016'],bottom = batsman_rec['2017'],label='2016')
plt.bar(batsman_rec['batsman'],batsman_rec['2015'],bottom = (batsman_rec['2016'] + batsman_rec['2017']),label='2015')
plt.xticks(batsman_rec['batsman'])
plt.legend()
plt.show()
````
